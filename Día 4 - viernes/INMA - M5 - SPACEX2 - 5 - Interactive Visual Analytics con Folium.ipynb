````python
# Crear objeto MarkerCluster
marker_cluster = MarkerCluster()

# Crear funci칩n para asignar color de marcador
def assign_marker_color(class_value):
    if class_value == 1:
        return 'green'
    else:
        return 'red'

# Crear nueva columna con los colores de los marcadores
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Crear marcadores para cada resultado de lanzamiento
for index, record in spacex_df.iterrows():
    coordenadas = [record['Lat'], record['Long']]
    marker = folium.Marker(coordenadas, icon=folium.Icon(color='white', icon_color=record['marker_color']))
    marker_cluster.add_child(marker)

# A침adir objeto MarkerCluster al objeto Map
site_map.add_child(marker_cluster)
````

# TAREA 3: Calcular las distancias entre un lugar de lanzamiento y sus proximidades


A continuaci칩n, tenemos que explorar y analizar las proximidades de los lugares de lanzamiento.


Primero vamos a a침adir un MousePosition en el mapa para obtener las coordenadas de un rat칩n sobre un punto del mapa. As칤, mientras exploras el mapa, puedes encontrar f치cilmente las coordenadas de cualquier punto de inter칠s (como el ferrocarril)

````python
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
````

Ahora acerque el zoom a un lugar de lanzamiento y explore su proximidad para ver si puede encontrar f치cilmente alguna v칤a f칠rrea, carretera, costa, etc. Mueve el rat칩n hacia estos puntos y marca sus coordenadas (mostradas en la parte superior derecha) para conocer la distancia al lugar de lanzamiento.


Puedes calcular la distancia entre dos puntos del mapa a partir de sus valores `Lat` y `Long` utilizando el siguiente m칠todo:


````python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # Radio de la tierra en km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
````

Selecci칩na un punto en la l칤nea de costa m치s cercana utilizando MousePosition para averiguar sus coordenadas. Calcula la distancia entre el punto de la l칤nea de costa y el lugar de lanzamiento.


````python
# Calcula la distancia a la coordenada de la costa m치s cercana
# e.g.,: Lat: 28.56367  Lon: -80.57163
ubi1 = [28.573255, -80.646895 ]
ubi2 = [46.921901599906725, 2.3292275138854777]
coordinates = [ubi1, ubi2]
distance_coastline = calculate_distance(coordinates[0][0], coordinates[0][1], coordinates[1][0], coordinates[1][1])
````

Para mostrar la linea que une los puntos utilizaremos el siguiente c칩digo:

````python
# Crear un objeto `folium.PolyLine` utilizando las coordenadas de la l칤nea de costa y la coordenada del lugar de lanzamiento
lines=folium.PolyLine(locations=coordinates, weight=1)
site_map.add_child(lines)
````

Despu칠s de obtener su coordenada, crea un `folium.Marker` para mostrar la distancia


````python
# Crear y a침adir un folio.Marcador en el punto de costa m치s cercano seleccionado en el mapa
# Muestra la distancia entre el punto de la l칤nea de costa y el lugar de lanzamiento utilizando la propiedad del icono 

distance_marker = folium.Marker(
   ubi2,
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
       )
   )
````
